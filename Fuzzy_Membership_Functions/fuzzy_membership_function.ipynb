```python
import numpy as np
import matplotlib.pyplot as plt
from math import e
```

## Traingular Membership Function

```python
def triangle(a, b, c):
    x_axis = []
    y_axis = []
    
    for i in np.arange(a, b, ((b-a)/10)):
        x_axis.append(i)
    for i in np.arange(b, c, ((c-b)/10)):
        x_axis.append(i)
    
    for no in x_axis:
        if a <= no and no < b:
            y_axis.append((no-a) / (b-a))
        elif b <= no and no < c:
            y_axis.append((c-no) / (c-b))
            
    x_axis.append(c)
    y_axis.append(0.0)
            
    # print(x_axis)
    # print(y_axis)
    
    plt.plot(x_axis, y_axis)
    plt.scatter(x_axis, y_axis)
    plt.plot([b,b], [0,1], 'r--')
    plt.title('Triangular Membership Function')
    plt.xlabel('x-axis')
    plt.ylabel('Membership Values')
    plt.show()
```

### Trapezoidal Membership Function

```python
def trapezoidal(a, b, c, d):
    x_axis = []
    y_axis = []
    
    for i in np.arange(a, b, ((b-a)/10)):
        x_axis.append(i)
    for i in np.arange(b, c, ((c-b)/10)):
        x_axis.append(i)
    for i in np.arange(c, d, ((d-c)/10)):
        x_axis.append(i)
    
    for no in x_axis:
        if a <= no and no < b:
            y_axis.append((no-a) / (b-a))
        elif b <= no and no < c:
            y_axis.append(1)
        elif c <= no and no < d:
            y_axis.append((d-no) / (d-c))
            
    x_axis.append(d)
    y_axis.append(0.0)
            
    # print(x_axis)
    # print(y_axis)
    
    plt.plot(x_axis, y_axis)
    plt.scatter(x_axis, y_axis)
    plt.plot([b,b], [0,1], 'r--')
    plt.plot([c,c], [0,1], 'r--')
    plt.title('Trapezoidal Membership Function')
    plt.xlabel('x-axis')
    plt.ylabel('Membership Values')
    plt.show()
```

### Gaussian Membership Function

```python
def gaussian(c, sd):
    x_axis = []
    y_axis = []
    
    for i in np.arange((c-c), ((c + c) + (((c + c) - c)/10)), (((c + c) - c)/10)):
        x_axis.append(float(i))
    
    for no in x_axis:
        ans = e ** ((-1/2) * (((no-c) / sd) ** 2))
        y_axis.append(float(ans))
            
    # print(x_axis)
    # print(y_axis)
    
    plt.plot(x_axis, y_axis)
    plt.scatter(x_axis, y_axis)
    plt.plot([c,c], [0,1], 'r--')
    plt.title('Gaussian Membership Function')
    plt.xlabel('x-axis')
    plt.ylabel('Membership Values')
    plt.show()
```

### Cauchy Membership Function

```python
def bell(a, b, c):
    x_axis = []
    y_axis = []
    
    for i in np.arange(a, c, 0.1):
        x_axis.append(i)
    for i in np.arange(c, c+(c-a)+0.1, 0.1):
        x_axis.append(i)
    
    for no in x_axis:
        ans = 1 / (1 + (abs((no-c) / a) ** (2*b)))
        y_axis.append(ans)            
            
    # print(x_axis)
    # print(y_axis)
    
    plt.plot(x_axis, y_axis)
    # plt.scatter(x_axis, y_axis)
    plt.plot([c,c], [0,1], 'r--')
    plt.title('Cauchy Membership Function')
    plt.xlabel('x-axis')
    plt.ylabel('Membership Values')
    plt.show()
```

### Sigmoidal Membership Function

```python
def sigmoid(a, c):
    x_axis = []
    y_axis = []
    
    for i in np.arange((c - (2*c)), c, 0.1):
        x_axis.append(i)
    for i in np.arange(c, (c + (2*c) + 0.1), 0.1):
        x_axis.append(i)
    
    for no in x_axis:
        ans = 1 / (1 + (e ** (-1 * ((no-c) / a))))
        y_axis.append(ans)            
            
    # print(x_axis)
    # print(y_axis)
    
    plt.plot(x_axis, y_axis)
    # plt.scatter(x_axis, y_axis)
    plt.plot([c,c], [0,1], 'r--')
    plt.title('Sigmoid Membership Function')
    plt.xlabel('x-axis')
    plt.ylabel('Membership Values')
    plt.show()
```
